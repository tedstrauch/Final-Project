### Keep only the columns we will be using for the analysis

### Bring in the Angela's child mortality data set

```python
child_mortality = pd.read_csv('C:/Users/singe/Downloads/U5MR_deaths.csv')
child_mortality.head()
```

### Rename some of the columns for ease of use and readability

```python
child_mortality.rename(columns={'CountryName' : 'Country', 'Uncertainty bounds*' : 'Uncertainty bounds', '2017.5' : 'Child_Mortality_2017'}, inplace=True)
child_mortality.head()
```

### Keep only the columns we need for analysis

```python
CM = child_mortality[[ 'Country', 'Uncertainty bounds', 'Child_Mortality_2017']]
CM.head()
```

### Select only the median value for child mortality rates. This gives us one country per row instead of 3. 

```python
CM4 = CM[child_mortality['Uncertainty bounds']=="Median"]
CM4.head()
```

### Merge the Happiness data set with the Child Mortality Rate data set. 

```python
Happy33 = pd.merge(HAPPY2017, CM4)
Happy33.head()
```

#### Check to see what types of data our columns are in. 

```python
Happy33.info()
```

#### Change Child Mortality 2017.5 from obfect to an int. in order to run correlation matrix and further analysis. 

```python
Happy33.Child_Mortality_2017 = Happy33.Child_Mortality_2017.str.replace(',', '').astype(int)
```
